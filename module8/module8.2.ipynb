```python
from datetime import datetime, timedelta
startDate = '2017-01-01'
endDate = '2017-01-07'
startDate_datatime = datetime.strptime(startDate, '%Y-%m-%d')
endDate_datatime = datetime.strptime(endDate, '%Y-%m-%d')
print(startDate_datatime, endDate_datatime)
```

```python
current_day = startDate_datatime


a = 1
while a < 5:
    print(a)
    a += 1

endDate_datatime -= timedelta(days = 4)
```

```python
current_day = startDate_datatime
while current_day <= endDate_datatime:
    print(current_day.strftime('%Y-%m-%d %H:%M:%S'))
    current_day += timedelta(hours = 1)
```
